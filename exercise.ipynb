## Từ mảng 1 chiều thành mảng 2 chiều 9 hàng 10 cột

```python
arr = np.arange(10, 100).reshape(9,10)
print(arr)
```

*np.eye(N, M=None, k=0)*  
- Tham số:
  - N: số hàng  
  - M: số cột (mặc định = N → ma trận vuông)  
  - k: đường chéo chính:  
      - 0: đường chéo chính  
      - \>0: dịch lên trên  
      - <0: dịch xuống dưới  

```python
arr = np.eye(N=8, dtype='int16')
print(arr)
```

```python
arr = np.array(
    [
      [
          [4, 5, 4, 2],
          [6, 3, 5, 1],
          [5, 2, 1, 2],
      ],
      [
          [7, 2, 3, 1],
          [6, 2, 0, 9],
          [0, 1, 9, 1],
      ]
    ]
)
print(arr)
```

```python
arr = np.array(
    [
        ['4', '3', '6', '2', '1'],
        ['7', '2', '1', '2', '6'],
    ],
    dtype='<U1',
)
print(arr)
arr = arr.astype('float16')
print(arr)
```
